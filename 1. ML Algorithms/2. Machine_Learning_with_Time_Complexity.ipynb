```python
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.neural_network import MLPClassifier
from concurrent.futures import ThreadPoolExecutor

# Function to generate synthetic data for testing
def generate_data(n_samples, n_features):
    X = np.random.rand(n_samples, n_features)
    y = np.random.randint(0, 2, size=n_samples)
    return X, y

# Function to measure training and inference time
def measure_time(algorithm, X, y, requires_no_target):
    training_times = []
    inference_times = []

    for _ in range(5):  # Run multiple iterations for averaging
        start_train = time.time()
        if requires_no_target:
            algorithm.fit(X)
        else:
            algorithm.fit(X, y)
        end_train = time.time()

        training_times.append(end_train - start_train)

        start_infer = time.time()
        if requires_no_target:
            algorithm.transform(X)
        else:
            algorithm.predict(X)
        end_infer = time.time()

        inference_times.append(end_infer - start_infer)

    return np.mean(training_times), np.mean(inference_times)

# Algorithms and their respective configurations
algorithms = {
    "Linear Regression": (LinearRegression(), False),
    "Logistic Regression": (LogisticRegression(), False),
    "Naive Bayes": (GaussianNB(), False),
    "Decision Tree": (DecisionTreeClassifier(), False),
    "Random Forest": (RandomForestClassifier(n_estimators=10), False),
    "Gradient Boosted Trees": (GradientBoostingClassifier(n_estimators=10), False),
    "Principal Component Analysis (PCA)": (PCA(n_components=5), True),
    "K-Nearest Neighbor": (KNeighborsClassifier(n_neighbors=5), False),
    "K-Means": (KMeans(n_clusters=3), True),
    "Dense Neural Networks": (MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=100), False)
}

# Function to execute timing for a specific algorithm
def execute_algorithm(name, algorithm, n_samples, n_features, requires_no_target):
    print(f"Running {name}...")
    X, y = generate_data(n_samples, n_features)

    if requires_no_target:
        y = None

    train_time, infer_time = measure_time(algorithm, X, y, requires_no_target)
    print(f"{name} - Training Time: {train_time:.4f}s, Inference Time: {infer_time:.4f}s")
    return name, train_time, infer_time

# Parameters for testing
n_samples_list = [100, 500, 1000, 5000]
n_features = 10
results = {}

# Multithreading to run algorithms in parallel
for n_samples in n_samples_list:
    with ThreadPoolExecutor() as executor:
        futures = [
            executor.submit(execute_algorithm, name, algo, n_samples, n_features, requires_no_target)
            for name, (algo, requires_no_target) in algorithms.items()
        ]

        for future in futures:
            name, train_time, infer_time = future.result()
            if name not in results:
                results[name] = {"train": [], "infer": []}
            results[name]["train"].append(train_time)
            results[name]["infer"].append(infer_time)

# Plotting results
for name, times in results.items():
    plt.figure()
    plt.plot(n_samples_list, times["train"], label="Training Time")
    plt.plot(n_samples_list, times["infer"], label="Inference Time")
    plt.title(f"{name} Time Complexity")
    plt.xlabel("Number of Samples")
    plt.ylabel("Time (s)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris, load_breast_cancer, load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
from concurrent.futures import ThreadPoolExecutor

# Predefined datasets for real-world testing
datasets = {
    "Iris": load_iris(return_X_y=True),
    "Breast Cancer": load_breast_cancer(return_X_y=True),
    "Digits": load_digits(return_X_y=True)
}

# Function to measure training and inference time and calculate metrics
def measure_time_and_metrics(algorithm, X_train, X_test, y_train, y_test, requires_no_target):
    training_times = []
    inference_times = []
    metrics = []

    for _ in range(5):  # Run multiple iterations for averaging
        start_train = time.time()
        if requires_no_target:
            algorithm.fit(X_train)
        else:
            algorithm.fit(X_train, y_train)
        end_train = time.time()

        training_times.append(end_train - start_train)

        start_infer = time.time()
        if requires_no_target:
            predictions = algorithm.transform(X_test) if hasattr(algorithm, 'transform') else algorithm.predict(X_test)
        else:
            predictions = algorithm.predict(X_test)
        end_infer = time.time()

        inference_times.append(end_infer - start_infer)

        # Calculate metrics
        if not requires_no_target:
            if isinstance(algorithm, (LinearRegression, LogisticRegression)):
                metric = r2_score(y_test, predictions) if y_test.ndim == 1 else mean_squared_error(y_test, predictions)
            else:
                metric = accuracy_score(y_test, np.round(predictions) if predictions.ndim == 1 else predictions.argmax(axis=1))
        else:
            metric = 0  # No meaningful metric for non-target-based algorithms in this setup

        metrics.append(metric)

    return np.mean(training_times), np.mean(inference_times), np.mean(metrics)

# Algorithms and their respective configurations
algorithms = {
    "Linear Regression": (LinearRegression(), False),
    "Logistic Regression": (LogisticRegression(max_iter=1000), False),
    "Naive Bayes": (GaussianNB(), False),
    "Decision Tree": (DecisionTreeClassifier(), False),
    "Random Forest": (RandomForestClassifier(n_estimators=10), False),
    "Gradient Boosted Trees": (GradientBoostingClassifier(n_estimators=10), False),
    "K-Nearest Neighbor": (KNeighborsClassifier(n_neighbors=5), False),
    "Dense Neural Networks": (MLPClassifier(hidden_layer_sizes=(50, 50), max_iter=500), False)
}

# Function to execute timing and metrics for a specific algorithm
def execute_algorithm(name, algorithm, dataset_name, X_train, X_test, y_train, y_test, requires_no_target):
    print(f"Running {name} on {dataset_name}...")

    train_time, infer_time, metric = measure_time_and_metrics(algorithm, X_train, X_test, y_train, y_test, requires_no_target)
    print(f"{name} - Dataset: {dataset_name}, Training Time: {train_time:.4f}s, Inference Time: {infer_time:.4f}s, Metric: {metric:.4f}")
    return name, train_time, infer_time, metric

# Parameters for testing
results = {}

# Multithreading to run algorithms in parallel
for dataset_name, (X, y) in datasets.items():
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    with ThreadPoolExecutor() as executor:
        futures = [
            executor.submit(execute_algorithm, name, algo, dataset_name, X_train, X_test, y_train, y_test, requires_no_target)
            for name, (algo, requires_no_target) in algorithms.items()
        ]

        for future in futures:
            name, train_time, infer_time, metric = future.result()
            if name not in results:
                results[name] = {"train": [], "infer": [], "metric": []}
            results[name]["train"].append(train_time)
            results[name]["infer"].append(infer_time)
            results[name]["metric"].append(metric)

# Plotting results
for name, times in results.items():
    plt.figure()
    plt.plot(list(datasets.keys()), times["train"], label="Training Time")
    plt.plot(list(datasets.keys()), times["infer"], label="Inference Time")
    plt.plot(list(datasets.keys()), times["metric"], label="Metric")
    plt.title(f"{name} Performance Metrics Across Datasets")
    plt.xlabel("Dataset")
    plt.ylabel("Values")
    plt.legend()
    plt.grid()
    plt.show()
```
